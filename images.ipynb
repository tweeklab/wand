```python
import os
from PIL import Image
import matplotlib as mpl
import matplotlib.pyplot as plt
import math

IMG_DIR = 'images/crap'
```

```python
glyph = "discard"
files = os.listdir(os.path.join(IMG_DIR, glyph))
numfiles = len(files)
images = []
for imfile in files:
    if imfile.endswith('.yaml'):
        continue
    images.append(Image.open(os.path.join(IMG_DIR, glyph, imfile)))

nrows = math.ceil(len(images)/10)
ncols = 10

fig = plt.figure(figsize=(ncols*1.5, nrows*1.5), dpi=75)

for i, (image, filename) in enumerate(zip(images, files)):
    ax = fig.add_subplot(nrows, ncols, i+1)
    ax.set_axis_off()
    ax.imshow(image, cmap=mpl.colormaps['gray'])
    ax.set_title(f"{filename}")
fig.tight_layout()
```
