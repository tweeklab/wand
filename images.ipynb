```python
import os
import numpy as np
from PIL import Image
import matplotlib as mpl
import matplotlib.pyplot as plt
import math

IMG_DIR = 'images/good'
```

```python
glyph = "tarantallegra"
files = os.listdir(os.path.join(IMG_DIR, glyph))
numfiles = len(files)
images = []
for imfile in files:
    if imfile.endswith('.yaml'):
        continue
    images.append(Image.open(os.path.join(IMG_DIR, glyph, imfile)))

nrows = math.ceil(len(images)) + 1
ncols = 1

fig = plt.figure(figsize=(ncols*1.5, nrows*1.5), dpi=75)

gradient = np.linspace(0, 255, 256)
gradient_image = np.vstack((gradient, gradient))

cmap_name = "hot"
ax = fig.add_subplot(nrows, ncols, 1)
ax.imshow(gradient_image, aspect='auto', cmap=mpl.colormaps[cmap_name])
ax.set_title("Colormap")

for i, (image, filename) in enumerate(zip(images, files)):
    ax = fig.add_subplot(nrows, ncols, i+2)
    ax.set_axis_off()
    ax.imshow(image, cmap=mpl.colormaps[cmap_name])
    ax.set_title(f"{filename}")
fig.tight_layout()

```
